### 14 Write the code to change each entry in the monthly_avg.month column so it shows the name of the month instead of the number. You must do this by creating a function that accepts the month (number) and returns the month name which you call from a map/lambda expression. A dictionary is useful here.

```python
def numToMon(n):
    monsDic={1:'January', 2:'February',3:"March",4:'April',
            5:'May',6:'June',7:'July',8:'August',9:'September',
            11:'November',12:'December',10:'October'}
    #Missed october but then went back for it, sorry for the weird order
    return monsDic[n]
```

```python
numToMon(12) #checking if works
```

```python
monthly_avg
```

```python
#dataframe['column to put new values in']=list(map(lambda x: functionToApply(x),dataframe['column for function to reference']))
```

```python
monthly_avg.month=list(map(lambda x: numToMon(x), monthly_avg['month']))
```

### 17 Write the code to change each entry in the hourly_avg index so it shows the name of the standard time (not military time). You must do this by creating a function that accepts the time (string) and returns the time in the format (12:00AM, 1:00AM, 2:00AM, . . . 10:00PM, 11:00PM) which you call from a map/lambda expression. A dictionary is useful here.

```python
hourly_avg.index
```

```python
def milToStanTime(n):
    hourDic={0:"12:00AM",  1:"1:00AM",  2:"2:00AM",
             3:"3:00AM",  4:"4:00AM",  5:"5:00AM",
             6:"6:00AM",  7:"7:00AM",  8:"8:00AM",
             9:"9:00AM",  10:"10:00AM",  11:"11:00AM",
             12:"12:00PM",13:"1:00PM",  14:"2:00PM",
             15:"3:00PM",  16:"4:00PM",  17:"5:00PM",
             18:"6:00PM",  19:"7:00PM",  20:"8:00PM",
             21:"9:00PM",  22:"10:00PM",  23:"11:00PM"}
    return hourDic[n]
```

```python
#dataframe['column to put new values in']=list(map(lambda x: functionToApply(x),dataframe['column for function to reference']))
```

```python
hourly_avg.index=list(map(lambda x: milToStanTime(x),hourly_avg.index))
```
